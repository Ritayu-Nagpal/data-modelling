#### Dataset

Predicting the profit of a startup company based on available metrics in the dataset

```python
# importing dataset
dataset = pd.read_csv('50_Startups.csv')
dataset.drop('Profit', axis=1).profile_report(title = 'Pandas Profiling', style={'full_width':True})
```

```python
train_X = dataset.iloc[:, :-1].values
train_Y = dataset.iloc[:, 4].values
```

Using Label Encoder to encode the categorical variable, Decision Tree and most tree based models do not perform better with using onehot encoding, however it can be still tried by uncommenting last 2 lines of code.

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
# categorical variable is at index 3
train_X[:, 3] = labelencoder.fit_transform(train_X[:, 3])
#onehotencoder = OneHotEncoder(categorical_features = [3])
#train_X = onehotencoder.fit_transform(train_X).toarray()
```

Splitting dataset into test and train dataset with 80/20 ratio

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(train_X, train_Y, test_size = 0.2, random_state = 0)
```

Fitting Decision Tree Regression model from Scikit-Learn library

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(X_train, Y_train)
```

```python
pd.DataFrame({'Feature':list(dataset.columns[:-1]), 'Feature Importance':list(tree_reg.feature_importances_)})
```

According to Decision Tree Model, R&D spend is the most important criteria to predict the profit of the company and Marketing Spend is the second most important criteria to predict profit

Predicting results on test datatest and scaling results back to original:

```python
Y_pred = tree_reg.predict(X_test)
```

```python
results = pd.DataFrame()
results['Actual'] = Y_test
results['Predicted'] = Y_pred
results
```

Calculating the R-square metric to check the correctness of the model:

```python
from sklearn.metrics import r2_score
model_score = r2_score(Y_test, Y_pred)
model_score
```
